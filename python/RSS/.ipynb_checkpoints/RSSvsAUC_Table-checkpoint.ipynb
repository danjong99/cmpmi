```python
import pandas as pd
from matplotlib import pyplot as pp
import numpy as np
from seaborn import distplot, heatmap, clustermap
```

```python
rss =pd.read_table( 'RSS.txt', sep=',', index_col=0)
```

```python
rss2 = rss.copy()
```

```python
rss2.head()
```

```python
for i,v in rss.iteritems() :
    rss2[i] = (v - np.mean(v))/ np.std(v)
```

```python
clustermap(rss2)
```

```python
for k,v in rss2.iteritems():
    distplot(v)
    pp.show()
    print( v[v.argsort()[-10:]] )
```

```python
auc = pd.read_table('./auc_mtx_avg10.txt', index_col=0, header=0 )
```

```python
auc.head()
```

```python
cell_info = pd.read_table( './cellinfo.txt' )
```

```python
cell_info.head()
```

```python
cell_info2 = cell_info[ cell_info.index.isin( auc.columns ) ]
```

```python
cell_info2.head()
```

```python
auc2 = pd.DataFrame( np.zeros( (auc.shape[0],13)  ), index=auc.index )
auc2.head()
```

```python
for k, v in auc2.iteritems():
    #print(k)
    #print(v)
    for regulon in v.index :
        #print(regulon)
        #break
        v[regulon] = auc.loc[ regulon, cell_info2[cell_info2.CellType == k].index ].mean()
    #break
```

```python
auc2 = pd.DataFrame( np.zeros( (auc.shape[0],13)  ), index=auc.index )
#auc2.head()
for k, v in auc2.iteritems():
    #print(k)
    #print(v)
    for regulon in v.index :
        #print(regulon)
        #break
        v[regulon] = auc.loc[ regulon, cell_info2[cell_info2.CellType == k].index ].mean()
    #break
```

```python
auc2.head()
```

```python
from seaborn import scatterplot

for k,v in rss2.iteritems() :
    w = auc2[int(k)]
    scatterplot( v, w )
    
    pp.xlabel( 'RSS-Z' )
    pp.ylabel( 'AUC' )
    pp.title( 'Cluster %s'%k )
    pp.show()
    
    vtop = v[ v.argsort()[-10:] ].index
    wtop = w[ w.argsort()[-10:] ].index
    total = set( list(vtop)+ list(wtop))
    vw = pd.DataFrame( {'RSS-Z':v[total], 'AUC':w[total] } )
    
    print( vw.iloc[ vw['RSS-Z'].argsort() ] )
```
